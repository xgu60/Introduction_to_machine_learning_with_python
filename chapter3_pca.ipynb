```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_blobs, load_breast_cancer
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

%matplotlib inline
```

```python
#generate data
X, y = make_blobs(n_samples=100, centers=20, random_state=0)
#scale up one feature
X[:, 0] = X[:, 0] * 100000000000000
```

```python
#apply pca 
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(8, 4))
axes[0].scatter(X[:, 0], X[:, 1])
axes[1].scatter(X_pca[:, 0], X_pca[:, 1]);
```

```python
#load breast cancer data
data = load_breast_cancer()
```

```python
X = data["data"]
y = data["target"]
```

```python
print(X.shape, y.shape)
```

```python
#standardize
X = StandardScaler().fit_transform(X)
```

```python
#apply pca
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
```

```python
#plot 
#left use first 2 features
#right use the 2 components generated from pca
fig, axes = plt.subplots(1, 2, figsize=(8, 4))
axes[0].scatter(X[y==0, 0], X[y==0, 1], c="blue")
axes[0].scatter(X[y==1, 0], X[y==1, 1], c="red")
axes[1].scatter(X_pca[y==0, 0], X_pca[y==0, 1], c="blue")
axes[1].scatter(X_pca[y==1, 0], X_pca[y==1, 1], c="red");
```
