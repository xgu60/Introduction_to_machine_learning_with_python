```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression

%matplotlib inline
```

```python
#load iris dataset
iris_dataset = load_iris()
```

```python
#check the type of iris_dataset
type(iris_dataset)
```

```python
#print keys of iris_dataset
iris_dataset.keys()
```

```python
#check the types
for k in iris_dataset.keys():
    print("type of " + k + ": {}".format(type(iris_dataset[k])))      
```

```python
print(iris_dataset["data"].shape)
print(iris_dataset["target"].shape)
print(iris_dataset["target_names"].shape)
```

```python
#create dataframe using data
data = pd.DataFrame(iris_dataset["data"], columns=iris_dataset["feature_names"])
```

```python
data.head()
```

```python
#pandas scatter matrix
pd.plotting.scatter_matrix(data, c=iris_dataset["target"], marker="o", s=20, alpha=0.8, figsize=(8, 8));
```

```python
#seaborn heatmap
fig, ax = plt.subplots(figsize=(5, 4))
sns.heatmap(data.corr(), ax=ax)
sns.despine()
```

```python
#target value counts
fig, ax = plt.subplots(figsize=(5, 3))
sns.countplot(iris_dataset["target"], ax=ax);
```

```python
#train and validation split
X_train, X_val, y_train, y_val = train_test_split(data, iris_dataset["target"], test_size=0.2, random_state=0)
```

```python
print(X_train.shape, X_val.shape)
```

```python
#train KNN model
knn = KNeighborsClassifier(n_neighbors=3)
```

```python
knn.fit(X_train, y_train)
```

```python
prediction = knn.predict(X_val)
print(prediction)
print(y_val)
print(np.mean(prediction == y_val))
```

```python
knn.score(X_val, y_val)
```

```python
lr = LogisticRegression(random_state=0, multi_class="ovr", solver="lbfgs")
lr.fit(X_train, y_train)
```

```python
prediction = lr.predict(X_val)
print(prediction)
print(y_val)
print(np.mean(prediction == y_val))
```
